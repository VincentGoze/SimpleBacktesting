## Implementation of some signals

Return best parameters, and then do the sum of signals to try new method</span>.


```python
def mean_reversion_signal(df,prices_column,window,threshold):
    """
    buy if price is lower than the mean of previous days (adjustable threshold)
    sell if price is higher 
    """
    means=df[prices_column].rolling(window).mean().shift(1)
    signal_column=[]
    for p,m in zip(df[prices_column], means):
        if p < (1-threshold)*m:
            signal_column.append(1)
        elif p > (1+threshold)*m:
            signal_column.append(-1)
        else:
            signal_column.append(0)
    return signal_column


def rsi_signal(df,prices_column,window,lower_threshold,upper_threshold):
    """
    RSI(Relative Strength Index)
    high value=too many buyers --> sell
    low value=too many sellers --> buy
    implemented with a simple moving average (not exponential)
    """

    #Average gains and losses
    delta = df[prices_column].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(window=window, min_periods=window).mean()
    avg_loss = loss.rolling(window=window, min_periods=window).mean()
    

    rsi = 100 - (100 / (1 + avg_gain / avg_loss))
    
    
    signal_column = []
    for val in rsi:
        if val < lower_threshold:
            signal_column.append(1)
        elif val > upper_threshold:
            signal_column.append(-1)
        else:
            signal_column.append(0)
    
    return signal_column

def moving_average_crossover_signal(df, prices_column, short_window, long_window, threshold):
    """
    Buy when the short average crosses above the long average.
    Sell when short avg. crosses below short avg.
    Use a threshold to avoid "noise"
    """
    short_avg = df[prices_column].rolling(short_window).mean().shift(1)
    long_avg = df[prices_column].rolling(long_window).mean().shift(1)
    
    signal_column = []
    
    for s,l in zip(short_avg,long_avg):
        if s > (1+threshold)*l:   
            signal_column.append(1)
        elif s < (1-threshold)*l: 
            signal_column.append(-1)
        else:
            signal_column.append(0)
    
    return signal_column
```

# Grid-Search type optimization

```python
def optimize_signal_parameters(signal_name,df,prices_column,parameters,pourcentages_buy,start_date="2020-01-01",initial_amount=100000,cutoff_threshold=2):
    """
    Given a signal_name function, test different parameters as well as different pourcentages allocation of portfolio when buying stocks on a dataframe df.
    The best results are plotted and compared to a Buy & Hold benchmark
    """

    # Buy & Hold benchmark
    plt.figure(figsize=(12, 6))
    df=df[df.index>=start_date]
    to_buy=math.floor(initial_amount/df.iloc[0]["Open"])
    cash_residual=initial_amount-to_buy*df.iloc[0]["Open"]
    buy_hold_value= cash_residual+to_buy*df["Open"]
    plt.plot(df.index, buy_hold_value, label='Buy & Hold', linestyle='--', color='black')


    best_parameters=0
    best_value=0
    
    
    #Parsing parameters and iterating over them
    parameters_keys = list(parameters.keys())
    for values in itertools.product(*parameters.values()):  
        for pourcentage in pourcentages_buy:
            
            cash=initial_amount
            n_shares=0
            
            signal=signal_name(df,prices_column,*values)
            nr_shares=[]
            total_value=[]
            for a in range(len(df)):
                row=df.iloc[a]
                sig=signal[a]
                if sig==1: #allocate pourcentage% of remaining cash to buy stocks
                    to_buy=math.floor(pourcentage*cash/row["Open"])
                    n_shares+=to_buy
                    cash-=(to_buy*row["Open"])
                elif sig==-1: #sell all acquired stocks
                    cash+=(n_shares*row["Open"])
                    n_shares=0
                nr_shares.append(n_shares)
                total_value.append(cash+n_shares*row["Open"])


            #Visualize only "good" results, those who multiply the initial_amount by at least cutoff_threshold
            if total_value[-1]>cutoff_threshold*initial_amount:
                print(f"{parameters_keys}={values}, Pourcentage={pourcentage}, Total Value={total_value[-1]:.2f}$")
                plt.plot(df.index, total_value, label=f'Parameters={values}')

            #Keeping track if best value
            if total_value[-1]>best_value:
                best_value=total_value[-1]
                best_parameters=dict(zip(parameters_keys,values))
                best_parameters["pourcentages_buy"]=pourcentage
            
    plt.show()

    #Return tracked best value
    return best_value, best_parameters
        
    
```

```python
best_value_mean_reversion, best_parameters_mean_reversion=optimize_signal_parameters(mean_reversion_signal,df,"Close",{"window":[4,10,15,20],"threshold":[0,0.02,0.05,0.08]},pourcentages_buy=[0.2,0.5,0.65,0.8],cutoff_threshold=1.3)

#The result with this signal is not very good
```
